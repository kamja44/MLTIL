```python
'''
파이썬 리스트
- 2개의 숫자와 하나의 문자열로 이뤄진 파이썬 리스트
- 리스트의 인덱스는 0부터 시작
'''
my_list = [10, "Hello list", 20]
print(my_list[1])
```

```python
# 리스트의 요소로 다른 리스트를 포함해도 된다.
# 이렇게 중첩된 배열을 2차원 배열이라고 한다.
# 파이썬 리스트로 3개의 요소를 가진 1차원 배열을 2개 쌓은 예
my_list_2 = [[10 ,20, 30], [40, 50, 60]] # 이 리스트의 크기를 2 x 3크기 라고 한다.
```

```python
# 2차원 배열의 요소를 선택할 때는 2개의 대괄호를 사용한다.
# 첫 번째 대괄호에서는 행의 인덱스를 넣고, 두 번째 대괄호에서는 열의 인덱스를 넣는다.
# 두 번째 행의 두 번째 열을 선택했기에, 50이 출력된다.
print(my_list_2[1][1])
```

```python
'''
넘파이를 이용하여 배열 생성
- 넘파이는 파이썬의 리스트처럼 숫자와 문자를 함께 담을 수 없다.
    - 딥러닝에서는 숫자 데이터만 다룬다.
        - 문자 데이터가 있어도 결국 숫자로 바꿔 입력해야 한다.
'''
import numpy as np

# 1. array() 함수로 2차원 배열 만들기
my_arr = np.array([[10, 20, 30], [40, 50, 60]])
print(my_arr)
```

```python
# 2. type() 함수로 넘파이 배열인지 확인하기
print(type(my_arr))
```

```python
# 3. numpy 배열에서 요소 선택하기
# 넘파이 배열의 요소를 선택하는 방법은 파이썬 리스트에서 요소를 선택하는 방법과 동일하다.
print(my_arr[0][2])
```

```python
# 4. numpy 내장 함수 사용하기
# np.sum() 함수를 사용하면 배열의 모든 요소를 더할 수 있다.
print(np.sum(my_arr))
```

```python
'''
matplotlib으로 그래프 그리기
'''
import matplotlib.pyplot as plt

'''
산점도
- x축을 기준으로 y축의 변화 추이를 살펴볼 때 사용한다.
'''

# 1. 선 그리기
# 선 그래프를 그리기 위해 x축의 값과 y축의 값을 matplotlib의 plot() 함수에 전달해야 한다.
# 그 뒤, show() 함수를 사용하여 화면에 그래프를 출력한다.
# 다음은 y = x^2 그래프를 그리는 예이다.
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25]) # x축, y축의 값을 파이썬 리스트로 전달한다.
plt.show()
```

```python

# 2. 산점도 그리기
# 데이터의 x축, y축 값을 이용하여 점으로 그래프를 그린다.
# 이때, show() 함수를 사용하지 않아도 matplotlib이 자동으로 그래프를 그려주지만, 이 함수를 사용하면 좀 더 깔끔하고 보기 좋은 그래프가 그려진다.
plt.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

```python
# 3. 넘파이 배열로 산점도 그리기
# 파이썬의 과학 패키지들은 예외 없이 넘파이 배열로 데이터를 주고받을 수 있게 만들어져있따.
# 여기서는 넘파이의 random.randn() 함수를 사용하여 표준 정규 분포를 따르는 난수(random number)를 만들고, 그 값을 이용하여 산점도를 그린다.
# 다음과 같이 입력하면 1000개의 난수를 만들어 산점도를 그릴 수 있다.
x = np.random.randn(1000) # 표준 정규 분포를 따르는 난수 1000개를 만든다.
y = np.random.randn(1000) # 표준 정규 분포를 따르는 난수 1000개를 만든다.
plt.scatter(x, y)
plt.show()
```
