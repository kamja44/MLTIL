```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2)
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

plt.figure(figsize = (16, 16))
tree.plot_tree(clf, feature_names = iris.feature_names, class_names= iris.target_names.tolist(), filled = True)
plt.show()
'''
petal width (cm) <= 0.8 
- 
gini
-  
samples
-   (120)
value
-  0  39
-  1  38
-  2  43
'''
```

```python
'''
max_depth   
'''
clf = DecisionTreeClassifier(max_depth= 3)
clf.fit(X_train, y_train)

plt.figure(figsize= (16, 16))
tree.plot_tree(clf, feature_names = iris.feature_names, class_names = iris.target_names.tolist(), filled = True)
```

```python
'''
min_samples_split 4 
-    
'''
clf = DecisionTreeClassifier(min_samples_split= 4)
clf.fit(X_train, y_train)

plt.figure(figsize= (16, 16))
tree.plot_tree(clf, feature_names = iris.feature_names, class_names = iris.target_names.tolist(), filled = True)
```

```python
'''
min_samples_leaf 4 
-      
'''
clf = DecisionTreeClassifier(min_samples_leaf= 4)
clf.fit(X_train, y_train)

plt.figure(figsize= (16, 16))
tree.plot_tree(clf, feature_names = iris.feature_names, class_names = iris.target_names.tolist(), filled = True)
```

```python
'''
Feature 

DecisionTreeClassifier
- feature_importances_  .
'''
print(clf.feature_importances_)
plt.barh(iris.feature_names, clf.feature_importances_)
```

```python
'''
  
'''
from sklearn.datasets import make_classification

X, y = make_classification(n_features= 2, n_redundant= 0, n_classes= 3, n_clusters_per_class= 1)
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
import numpy as np

def visualize (model, X, y):
    X1 = X[:, 0]
    X2 = X[:, 1]
    plt.scatter(X1, X2, c = y)
    x1_min = np.min(X1)
    x1_max = np.max(X1)
    x2_min = np.min(X2)
    x2_max = np.max(X2)
    xx, yy = np.meshgrid(np.linspace(x1_min, x1_max, num = 200), np.linspace(x2_min, x2_max, num = 200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    n_classes = len(np.unique(y))
    plt.contourf(xx, yy, Z, alpha = 0.3, cmap = "rainbow")
clf = DecisionTreeClassifier()
clf.fit(X, y)
visualize(clf, X, y)
```

```python
clf = DecisionTreeClassifier(min_samples_leaf=6)
clf.fit(X, y)
visualize(clf, X, y)
```

```python
'''
  (UCL  )
'''
import pandas as pd
df = pd.read_csv("D:// /uci/features.txt", sep = "\s+", header = None, names = ["column_index", "column_name"])
df
```

```python
'''
  
'''
dup = df.groupby("column_name").count()
print(dup[dup["column_index"] > 1].count())
dup[dup["column_index"] > 1].head()
```

```python
'''
  
'''
df2 = pd.DataFrame(data = df.groupby("column_name").cumcount(), columns= ["cnt"])
df2 = df2.reset_index()
df3 = pd.merge(df.reset_index(), df2, how = "outer")
df3["column_name"] = df3[["column_name", "cnt"]].apply(lambda x: x[0]+"_"+str(x[1]) if x[1] > 0 else x[0], axis = 1)
df3 = df3.drop(["index"], axis=1)
feature_name = df3.iloc[:, 1].values.tolist()
```

```python
'''
 
'''
X_train = pd.read_csv("D:// /uci/train/X_train.txt", sep="\s+", names=feature_name)
X_test = pd.read_csv("D:// /uci/test/X_test.txt", sep="\s+", names=feature_name)
y_train = pd.read_csv("D:// /uci/train/y_train.txt", sep="\s+", header = None, names = ["action"])
y_test = pd.read_csv("D:// /uci/test/y_test.txt", sep="\s+", header = None, names = ["action"])
```

```python
print(X_train.info())
```

```python
print(y_train["action"].value_counts()) # action => label
```

```python
'''
  
'''
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_hat = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_hat)
print(accuracy)

print(clf.get_params()) #   
```

```python
'''
    
'''
from sklearn.model_selection import GridSearchCV

params = {"max_depth" : [6, 8, 10, 12, 16, 20, 24]}
cv = GridSearchCV(clf, param_grid= params, scoring= "accuracy", cv = 5)
cv.fit(X_train, y_train)
print(cv.best_score_)
print(cv.best_params_)
```

```python
results = pd.DataFrame(cv.cv_results_)
results[["param_max_depth", "mean_test_score"]]
```

```python
params = {"max_depth" : [6, 8, 10, 12, 16, 20, 24], "min_samples_split" : [16 , 24]} # min_samples_split 
cv = GridSearchCV(clf, param_grid= params, scoring= "accuracy", cv = 5)
cv.fit(X_train, y_train)
print(cv.best_score_)
print(cv.best_params_)
```

```python
best_clf = cv.best_estimator_
y_hat = best_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_hat)
print(accuracy)
```
